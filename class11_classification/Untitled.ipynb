```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.DataFrame(np.random.randint(80,100, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s2 = pd.DataFrame(np.random.randint(70,79, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s3 = pd.DataFrame(np.random.randint(60,69, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s4 = pd.DataFrame(np.random.randint(50,59, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s5 = pd.DataFrame(np.random.randint(40,49, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s6 = pd.DataFrame(np.random.randint(33,39, (10000,5)), columns=["s1","s2","s3","s4","s5"])
s7 = pd.DataFrame(np.random.randint(0,32, (10000,5)), columns=["s1","s2","s3","s4","s5"])

df = pd.concat([s1,s2,s3,s4,s5,s6,s7])
df.head()
```

# Symbolic AI

```python
def grade(sb1,sb2,sb3,sb4,sb5):
    total = sb1+sb2+sb3+sb4+sb5
    per = total/500*100
    grade = "Null"
    if per>=80:
        grade = "A+"
    elif per>=70:
        grade ="A"
    elif per>=60:
        grade = "B"
    elif per>=50:
        grade = "C"
    elif per>=40:
        grade = "D"
    elif per>=33:
        grade = "E"
    elif per>=0:
        grade = "Fail"
    
    return grade, 500
    
    
df['grade'] = df.loc[:,"s1":"s5"].apply(lambda x:grade(*x), axis=1)
df.grade.value_counts()
```

```python
def grade(sb1,sb2,sb3,sb4,sb5):
    obtained = sb1+sb2+sb3+sb4+sb5
    per = obtained/500*100
    grade = "Null"
    if per>=80:
        grade = "A+"
    elif per>=70:
        grade ="A"
    elif per>=60:
        grade = "B"
    elif per>=50:
        grade = "C"
    elif per>=40:
        grade = "D"
    elif per>=33:
        grade = "E"
    elif per>=0:
        grade = "Fail"
    
    return grade, 500, obtained, per
    
    
```

```python
    
df[["grade","total","obtained",'percentage']] = df.loc[:,"s1":"s5"].apply(lambda x:grade(*x), axis=1, result_type='expand' )
df.head()
```

# create data for ML and DL

```python
data = df[["s1","s2","s3","s4","s5",'grade']].copy()
data
```

### one-hot encoding

```python
X = data.loc[:,"s1":"s5"] # input
X
```

```python
y = data['grade']
y
```

#### convert into onehotencode

```python
y = pd.get_dummies(y)
y
```

```python
y.drop_duplicates()
```

# Deep learning

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, SGD
```

# Create model

```python
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(5,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(7, activation='softmax'))# 7 number of classes, activation directly relate with problem
```

```python
model.summary()
```

# Compilation process

```python
model.compile(optimizer=SGD(learning_rate=0.002),
             loss='categorical_crossentropy',
             metrics=['acc'])
```

# Now Start Training

```python
model.fit(X,y, epochs=70)
```
